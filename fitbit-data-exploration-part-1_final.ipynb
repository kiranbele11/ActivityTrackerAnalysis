# Question 3 - Visualization
### 3.1 Create visualizations to show steps of all persons involved over time

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df2 = df1.groupby(['ActivityDate'])['TotalSteps'].sum()
df2 = pd.DataFrame(df2)
df2.reset_index(inplace = True)

xpoints = df2.ActivityDate
ypoints = df2.TotalSteps

plt.plot(xpoints, ypoints)

plt.xticks(rotation=70)
plt.xlabel('Activity Date')
plt.ylabel('Total Steps')

plt.title('Total Steps Over Time')

plt.show()
```

### 3.5  Can you generate a heatmap that visualizes the correlation coefficients between the numerical columns like TotalSteps, TotalDistance, VeryActiveMinutes, LightActiveMinutes, and Calories?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Select the numerical columns of interest
numerical_columns = ['TotalSteps', 'TotalDistance', 'VeryActiveMinutes', 'LightlyActiveMinutes', 'Calories']

# Calculate the correlation matrix
correlation_matrix = df1[numerical_columns].corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')

# Show the plot
plt.tight_layout()
plt.show()
```

### 3.6 Can you draw a box plot for the TotalSteps to identify the median, quartiles, and any potential outliers in the dataset? Please label the y-axis appropriately.

```python
import pandas as pd
import matplotlib.pyplot as plt

total_steps_data = df1['TotalSteps']

# Create a box plot
plt.figure(figsize=(8, 6))
plt.boxplot(total_steps_data, vert=False)
plt.title('Box Plot of TotalSteps')
plt.xlabel('TotalSteps')
plt.ylabel('')

# Show the plot
plt.tight_layout()
plt.show()
```
